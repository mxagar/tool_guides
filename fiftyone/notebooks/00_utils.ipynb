```python
import numpy as np
import pandas as pd
import ast
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../datasets/vectors_dataset.csv")
df['embedding'] = df['embedding'].apply(ast.literal_eval)
```

```python
df.head()
```

```python
# Which the class distribution among clusters?
for cluster in df.cluster.unique():
    plt.figure(figsize=(3,3))
    df[df.cluster == cluster]["label"].hist()
    candidate_label = df[df.cluster == cluster]["label"].value_counts().index[0]
    plt.title(f"Cluster: {cluster} - {candidate_label}")
    plt.show()
```

```python
# Which the cluster distribution among labels?
cluster_pred = {}
for label in df.label.unique():
    plt.figure(figsize=(3,3))
    df[df.label == label]["cluster"].hist()
    candidate_cluster = df[df.label == label]["cluster"].value_counts().index[0]
    cluster_pred[candidate_cluster] = label
    plt.title(f"Label: {label} - {candidate_cluster}")
    plt.show()
```
