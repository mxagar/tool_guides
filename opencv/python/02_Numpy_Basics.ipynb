### 2.2. Loading Images

```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pic = Image.open('../data/00-puppy.jpg')
```

```python
type(pic)
```

```python
plt.imshow(pic)
```

```python
im = np.asarray(pic)
```

```python
plt.imshow(im)
```

```python
im.shape
```

```python
im_channel = im.copy()
```

```python
im_channel = im_channel[:,:,0]
```

```python
plt.imshow(im_channel, cmap='gray')
```

```python
im_channel = im.copy()
```

```python
im_channel[:,:,1] = 0
```

```python
plt.imshow(im_channel)
```
